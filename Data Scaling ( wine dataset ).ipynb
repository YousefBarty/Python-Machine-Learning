```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/'
                      'ml/machine-learning-databases/wine/wine.data',
                      header=None)

df_wine.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash',
                   'Alcalinity of ash', 'Magnesium', 'Total phenols',
                   'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                   'Color intensity', 'Hue', 'OD280/OD315 of diluted wines',
                   'Proline']

print('Class labels', np.unique(df_wine['Class label']))
df_wine.head()
```

```python
# df_wine = pd.read_csv('https://raw.githubusercontent.com/rasbt/python-machine-learning-book/master/code/datasets/wine/wine.data', header=None)
```

```python
y=df_wine['Class label']
y.head()
```

```python
X=df_wine.iloc[:,1:]
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr=LogisticRegression(solver='lbfgs',C=10)
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

## Bringing features onto the same scale

Whether to normalize data before or after splitting depends on the specific context and the goals of your machine learning model. However, a common practice is to perform normalization after splitting the data. Here's why:

Information Leakage: Normalizing the entire dataset before splitting can lead to information leakage from the test set into the training set. The mean and standard deviation calculated during normalization are based on the entire dataset, including both training and test sets. This can introduce a form of data leakage because the model is exposed to information from the test set during training.

Real-world Scenario Simulation: When you train a machine learning model, it should be able to generalize well to unseen data. Splitting the data into training and test sets simulates a real-world scenario where the model encounters new, unseen data. Normalizing after splitting ensures that the normalization parameters are computed based only on the training data, allowing the model to generalize better to new, unseen data.

Here's a typical sequence of steps:

Split the Data: Divide your dataset into training and test sets.
Calculate Normalization Parameters: Calculate the mean and standard deviation (or other normalization parameters) based only on the training set.
Normalize the Data: Apply normalization to both the training and test sets using the parameters calculated from the training set.
This approach helps ensure that your model's performance is accurately assessed on truly unseen data, and it reduces the risk of overfitting to the peculiarities of the test set.


```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train_norm = mms.fit_transform(X_train) 
X_test_norm = mms.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr1=LogisticRegression(solver='lbfgs',C=10)
lr1.fit(X_train_norm,y_train)
y_pred = lr1.predict(X_test_norm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr2=LogisticRegression(solver='lbfgs',C=10)
lr2.fit(X_train_std,y_train)
y_pred = lr2.predict(X_test_std)
print(accuracy_score(y_test,y_pred))
```

```python
ex = pd.DataFrame([0, 1, 2, 3, 4, 5])

# standardize
ex[1] = (ex[0] - ex[0].mean()) / ex[0].std(ddof=0)

# Please note that pandas uses ddof=1 (sample standard deviation) 
# by default, whereas NumPy's std method and the StandardScaler
# uses ddof=0 (population standard deviation)

# normalize
ex[2] = (ex[0] - ex[0].min()) / (ex[0].max() - ex[0].min())
ex.columns = ['input', 'standardized', 'normalized']
ex
```

## Assessing Feature Importances with Random Forests

```python
import matplotlib.pyplot as plt 
```

```python
from sklearn.ensemble import RandomForestClassifier

feat_labels = df_wine.columns[1:]

forest = RandomForestClassifier(n_estimators=10000,
                                random_state=0,
                                n_jobs=-1)

forest.fit(X_train, y_train)
importances = forest.feature_importances_

indices = np.argsort(importances)[::-1]

for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, 
                            feat_labels[indices[f]], 
                            importances[indices[f]]))

plt.title('Feature Importances')
plt.bar(range(X_train.shape[1]), 
        importances[indices],
        color='lightblue', 
        align='center')

plt.xticks(range(X_train.shape[1]), 
           feat_labels[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
#plt.savefig('./random_forest.png', dpi=300)
plt.show()
```
